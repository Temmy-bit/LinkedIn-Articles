```python
from scipy.stats import boxcox
import numpy as np

# Example skewed data with outliers
data = np.array([1, 2, 3, 4, 100, 120, 150])  # Right-skewed

# Log Transformation
log_transformed = np.log(data + 1)  # Adding 1 to avoid log(0)

# Box-Cox Transformation (only works for positive data)
boxcox_transformed, lambda_opt = boxcox(data + 1)

print("Log Transformed Data:", log_transformed)
print("Box-Cox Transformed Data:", boxcox_transformed)
print("Optimal λ:", lambda_opt)
```

```python
from sklearn.cluster import DBSCAN
import numpy as np
import matplotlib.pyplot as plt

# Sample dataset
X = np.array([[1, 2], [2, 2], [2, 3], [8, 7], [8, 8], [25, 80]])  # Last point is an outlier

# Apply DBSCAN
db = DBSCAN(eps=2, min_samples=2).fit(X)
labels = db.labels_

# Visualizing clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', edgecolors='k')
plt.title("DBSCAN Clustering and Outlier Detection")
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats.mstats import winsorize
from scipy.stats import boxcox

# Sample data with outliers
data = np.array([1, 5, 10, 50, 100, 500, 1000])

# Winsorization (5% capping)
winsorized_data = winsorize(data, limits=[0.05, 0.05])

# Log Transformation
log_transformed = np.log1p(data)  # log1p avoids log(0) issues

# Box-Cox Transformation
boxcox_transformed, best_lambda = boxcox(data)

# Capping (Truncation)
lower_bound = np.percentile(data, 5)  # 5th percentile
upper_bound = np.percentile(data, 95)  # 95th percentile
capped_data = np.clip(data, lower_bound, upper_bound)

# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Winsorization Plot
sns.boxplot(data=[data, winsorized_data], showfliers=True, ax=axes[0, 0])
axes[0, 0].set_xticklabels(['Original', 'Winsorized'])
axes[0, 0].set_title("Winsorization")

# Log Transformation Plot
sns.histplot(data, kde=True, label="Original", color="red", alpha=0.5, ax=axes[0, 1])
sns.histplot(log_transformed, kde=True, label="Log Transformed", color="blue", alpha=0.5, ax=axes[0, 1])
axes[0, 1].legend()
axes[0, 1].set_title("Log Transformation")

# Box-Cox Transformation Plot
sns.histplot(data, kde=True, label="Original", color="red", alpha=0.5, ax=axes[1, 0])
sns.histplot(boxcox_transformed, kde=True, label=f"Box-Cox (λ={round(best_lambda, 2)})", color="blue", alpha=0.5, ax=axes[1, 0])
axes[1, 0].legend()
axes[1, 0].set_title("Box-Cox Transformation")

# Capping (Truncation) Plot
sns.boxplot(data=[data, capped_data], showfliers=True, ax=axes[1, 1])
axes[1, 1].set_xticklabels(['Original', 'Capped'])
axes[1, 1].set_title("Capping (Truncation)")

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from scipy.stats import zscore

# Set a random seed for reproducibility
np.random.seed(42)

# Generate Univariate Data with Outliers
data_univariate = np.random.normal(50, 10, 100)  # Normal distribution (mean=50, std=10)
outliers_univariate = np.array([10, 100])  # Extreme values
data_univariate = np.append(data_univariate, outliers_univariate)

# Generate Multivariate Data with Outliers
X, _ = make_blobs(n_samples=100, centers=[[2, 2], [5, 5]], cluster_std=0.8, random_state=42)
outliers_multivariate = np.array([[10, 10], [12, 12]])  # Extreme points
X = np.vstack([X, outliers_multivariate])

# Plot Univariate Outliers
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.boxplot(x=data_univariate, color='skyblue')
plt.title("Univariate Outliers (Box Plot)");

# Plot Multivariate Outliers
plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], label="Normal Data", alpha=0.7)
plt.scatter(outliers_multivariate[:, 0], outliers_multivariate[:, 1], color='red', label="Outliers", s=100, edgecolors='black')
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.title("Multivariate Outliers (Scatter Plot)")

plt.tight_layout()
plt.show()
```
